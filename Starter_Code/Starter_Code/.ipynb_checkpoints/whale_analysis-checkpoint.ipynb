## S&P 500 Returns

Read the S&P 500 historic closing prices and create a new daily returns DataFrame from the data. 

```python
# Reading S&P 500 Closing Prices
sp500path = Path(r"C:\Users\gamez\Downloads\Starter_Code\Starter_Code\Resources\sp500_history.csv")
sp500_returns = pd.read_csv(sp500path)
sp500_returns.set_index(pd.to_datetime(sp500_returns['Date'], infer_datetime_format=True), inplace=True)

sp500_returns.head()
```

```python
# Check Data Types
sp500_returns.dtypes
```

```python
# Fix Data Types
sp500_returns["Close"] = sp500_returns["Close"].str.replace("$", "")
sp500_returns['Close'] = sp500_returns['Close'].astype('float')
sp500_returns.dtypes
```

```python
sp500_returns.drop(columns=['Date'], inplace=True)
sp500_returns.head()
```

```python
# Calculate Daily Returns
daily_returns = sp500_returns.pct_change()
daily_returns
```

```python
daily_returns.loc['2015-03-03']
```

```python
# Rename `Close` Column to be specific to this portfolio.
daily_returns.rename(columns={'Close':'SP500 Daily Returns'},inplace=True)
daily_returns.sort_values('Date', ascending=True)
```

```python
daily_returns.loc['2015-03-03']
```

```python
# Join Whale Returns, Algorithmic Returns, and the S&P 500 Returns into a single DataFrame with columns for each portfolio's returns.
appended_data = pd.concat([whale_returns, algorithmic_returns,daily_returns], axis="columns")
appended_data.head(500)
```

### Calculate Standard Deviations

```python
# Calculate the daily standard deviations of all portfolios
cumulative_returns.std().plot()
plt.xticks(rotation=90)
```

### Determine which portfolios are riskier than the S&P 500

```python
# Calculate  the daily standard deviation of S&P 500
print(f"The standard deviation of S&P500 is:",cumulative_returns['SP500 Daily Returns'].std())


# Determine which portfolios are riskier than the S&P 500

print(f"The standard deviation of Soros is:",cumulative_returns['SOROS FUND MANAGEMENT LLC'].std())
print(f"The standard deviation of Paulson is:",cumulative_returns['PAULSON & CO.INC. '].std())
print(f"The standard deviation of Tiger is:",cumulative_returns['TIGER GLOBAL MANAGEMENT LLC'].std())
print(f"The standard deviation of Bershire is:",cumulative_returns['BERKSHIRE HATHAWAY INC'].std())
print(f"The standard deviation of Algo1 is:",cumulative_returns['Algo 1'].std())
print(f"The standard deviation of Algo2 is:",cumulative_returns['Algo 2'].std())

print(f"")

print("------" *10)

#Tried to create a loop that will go through the cumulative returns and separate the ones with higher and the ones with lower standard deviation (I was not able to do it successfully) 
risk_status_vs_sp500 = {}
risk_status_vs_sp500["higher"] = []
risk_status_vs_sp500["lower"]= []

    
for x in cumulative_returns.std():
    if (x) > float(cumulative_returns['SP500 Daily Returns'].std()):
        risk_status_vs_sp500["higher"].append(x)
    elif x == float(cumulative_returns['SP500 Daily Returns'].std()):
        print(f"SP500 standard deviation is:",cumulative_returns['SP500 Daily Returns'].std())
    else:
        risk_status_vs_sp500["lower"].append(x)
        
print(f"The higher devations are:",risk_status_vs_sp500["higher"])
print(f"The lower devations are:",risk_status_vs_sp500["lower"])

# According to investopedia, standard deviation tells the trading consistancy of a fund. A high standard deviation means greater volatility and viceversa with a low standar deviation. 
#source: https://www.investopedia.com/ask/answers/022015/what-does-standard-deviation-measure-portfolio.asp
```

### Calculate the Annualized Standard Deviation

```python
# Calculate the annualized standard deviation (252 trading days)
from math import sqrt
squareroot252 = sqrt(252)
annualizedstd = squareroot252 * cumulative_returns.std()
annualizedstd.plot()
plt.xticks(rotation=90)

#Resource for annualized standard deviation = "https://financetrain.com/calculate-annualized-standard-deviation"
```

## Rolling Statistics

Risk changes over time. Analyze the rolling statistics for Risk and Beta. 

1. Calculate and plot the rolling standard deviation for all portfolios using a 21-day window
2. Calculate the correlation between each stock to determine which portfolios may mimick the S&P 500
3. Choose one portfolio, then calculate and plot the 60-day rolling beta between it and the S&P 500

### Calculate and plot rolling `std` for all portfolios with 21-day window

```python
# Calculate the rolling standard deviation for all portfolios using a 21-day window
cumulative_returns.rolling(window=21).std()
# Plot the rolling standard deviation
cumulative_returns.rolling(window=21).std().plot()
```

### Calculate and plot the correlation

```python
import seaborn as sns
%matplotlib inline
# Calculate the correlation
price_correlation = cumulative_returns.corr()
# Display de correlation matrix
sns.heatmap(price_correlation, vmin=-1, vmax=1)
```

### Calculate and Plot Beta for a chosen portfolio and the S&P 500

```python
# Calculate covariance of a single portfolio
# Chosen Portfolio : 'PAULSON & CO.INC. '
covariance_paulson = cumulative_returns['PAULSON & CO.INC. '].cov(cumulative_returns['PAULSON & CO.INC. '])
print(f"The covariance of PAULSON & CO.INC. LLC is:",covariance_paulson)
# Calculate variance of S&P 500
variance_SP500 = cumulative_returns['SP500 Daily Returns'].var()
print(f"Variance of S&P 500 is:",variance_SP500)
# Computing beta
beta = covariance / variance
print(f"beta between PAULSON & CO and the SP500 is:",beta)

# Plot beta trend ( only plotting covariance and variance to be able to see difference in new chart below)
rolling_covariance = cumulative_returns['PAULSON & CO.INC. '].rolling(window=60).cov(cumulative_returns['SP500 Daily Returns'])
rolling_covariance.plot(figsize=(50, 10))
                        
rolling_variance = cumulative_returns['SP500 Daily Returns'].rolling(window=60).var()
rolling_variance.plot(figsize=(50, 10))

# Plot beta trend (did not use- to be able to see difference in the following chart)                   
#rolling_beta = rolling_covariance / rolling_variance
#rolling_beta.plot(figsize=(20, 10), title='Rolling 60-Day Beta of Tiger')
```

```python
# Plot beta trend
rolling_beta = rolling_covariance / rolling_variance
rolling_beta.plot(figsize=(50, 10), title='Rolling 60-Day Beta of PAULSON')
```
